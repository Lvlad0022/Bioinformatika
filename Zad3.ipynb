```python
import random as rn


f1=open("acids.txt", "r")
ak=f1.readline()
f1.close()

bm=[]
f1=open("blosum50.txt", "r")
for i in range(20):
    line=f1.readline()
    vc=line.split()
    bm.append(vc[:])
for i in range(20):
    for j in range(20):
        bm[i][j]=int(bm[i][j])
f1.close()
```

```python
def score (x,y):

    m=len(x)
    n=len(y)

    sm=[] 
    tmp=[]
    for i in range(m+1):
        tmp.append(0)
    for i in range(n+1):
        sm.append(tmp[:])


    for i in range(n+1):
        sm[i][0]=max(0,-8*i)
    for i in range(m+1):
        sm[0][i]=max(0,-8*i)

    for i in range(1,n+1):
        for j in range(1,m+1):
            tmp=[]
            tmp.append(max(0,sm[i-1][j]-8))
            tmp.append(max(0,sm[i][j-1]-8))
            bb=bm[ak.index(y[i-1])][ak.index(x[j-1])]
            tmp.append(max(0,sm[i-1][j-1]+bb))
            tmp.sort()
            sm[i][j]=tmp[2]
    return max(map(max, sm))
```

```python
N = 10000
niz = 10
scorovi = []


for i in range(N):
    x = ''.join((rn.choices(ak,k=niz)))
    y = ''.join((rn.choices(ak,k=niz)))
    scorovi.append(score(x,y))

```

```python
import  matplotlib.pyplot as pl

pl.hist(scorovi)
pl.show()
```

```python
N = 10000
niz = 10
glave = []


for i in range(N):
    L = rn.choices([0,1],k=niz)
    MG = 0
    G=0
    for i in L:
        if i == 1:
            G = G+1
            if G>MG :
                MG=G
        else:
            G=0
    glave.append(MG)
```

```python
import  matplotlib.pyplot as pl

pl.hist(glave)
pl.show()
```
